```python
import numpy as np
import pandas as pd
```

```python
felder = pd.read_csv(".\Eigenschaften.csv")
for i in felder.index:
    felder.loc[i,"Eigenschaften"] = felder["Eigenschaften"][i].replace('-', ',')
felder
```

```python
def map(i):
    return felder.iloc[i,0]

def do_map(arr):
    return np.array([map(i) for i in arr])
```

```python
pd.options.display.max_colwidth = 500
```

```python
num_of_dif_bingos = 28
bingos = []
for i in range(num_of_dif_bingos):
    rnd = np.random.choice(range(43), 16, replace=False)
    tasks = do_map(rnd).reshape((4,4))
    bingos.append(pd.DataFrame(tasks))
```

```python
existing_sheet = pd.read_excel("output.xlsx")
```

```python
for i in range(num_of_dif_bingos):
    # bingos[i].to_excel("output.xlsx", sheet_name=f'sheet{i+1}', startrow=i, header=False, index=False)
    with pd.ExcelWriter('output.xlsx', engine='openpyxl', mode='a') as writer:
        bingos[i].to_excel(writer, sheet_name=f'Sheet{i}', index=False, header=None)
```
